```python
import numpy as np
import pandas as pd

df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

X = df[["Age"]]
y = df[["Spending Score (1-100)"]]
plt.xlabel("annual income")
plt.ylabel("spending score")
plt.scatter(X, y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, random_state=21, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

poly_deg = PolynomialFeatures(degree=4)
X_poly = poly_deg.fit_transform(X_train)

model = LinearRegression()
model.fit(X_poly, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

poly_deg = PolynomialFeatures(degree=4)
X_test = poly_deg.fit_transform(X_test)
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)
mse = mean_squared_error(y_true=y_test, y_pred=y_pred)
rmse = mean_squared_error(y_true=y_test, y_pred=y_pred, squared=False)
```

```python
print(mae)
print(mse)
print(rmse)
```

```python
import matplotlib.pyplot as plt

X = df[["Age"]]
y = df[["Spending Score (1-100)"]]
plt.xlabel("annual income")
plt.ylabel("spending score")
plt.scatter(X, y)

X_poly = poly_deg.fit_transform(X)
plt.plot(X_poly, model.predict(X_poly), color='red')
```
