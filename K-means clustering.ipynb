```python
import pandas as pd
import numpy as np
import math
  
# creating the Numpy array
array = np.array([
    [1, 1, 0], 
    [2, 4, 0], 
    [3, 9, 1], 
    [4, 6, 1], 
    [5, 5, 1], 
    [6, 3, 0],              
    [7, 4, 1],
    [8, 9, 1],
    [1, 5, 0], 
    [4, 3, 0],              
    [7, 1, 0],
    [3, 9, 1],
    [0, 0, 0],
    [1, 0, 0],
    [2, 2, 0],
])
   
# creating a list of column names
column_values = ['x', 'y', 'sum>10']
  
# creating the dataframe
df = pd.DataFrame(
    data = array, 
    columns = column_values
)
  
# displaying the dataframe
df
```

```python
# Setting up

K = 2

C1 = np.array([0])
C2 = np.array([1])

new_C1 = np.array([]).astype(int)
new_C2 = np.array([]).astype(int)
```

```python
# function to generate distances

def get_Distances(df):
    C1_x_sum = 0
    C1_y_sum = 0

    for o in C1:
        C1_x_sum += df.loc[df.index[o], 'x']
        C1_y_sum += df.loc[df.index[o], 'y']

    C2_x_sum = 0
    C2_y_sum = 0

    for o in C2:
        C2_x_sum += df.loc[df.index[o], 'x']
        C2_y_sum += df.loc[df.index[o], 'y']

    df['C1_distance^2'] = (df['x'] - (C1_x_sum / len(C1)))**2 + (df['y'] - (C1_y_sum / len(C1)))**2
    df['C2_distance^2'] = (df['x'] - (C2_x_sum / len(C2)))**2 + (df['y'] - (C2_y_sum / len(C2)))**2
    df.head(10)
    
    return df
```

```python
# Loop till concurrency

while (not np.array_equal(C1, new_C1) and not np.array_equal(C2, new_C2)):
    
    df = get_Distances(df)
    
    new_C1 = np.array([]).astype(int)
    new_C2 = np.array([]).astype(int)

    for index in range(len(df)):
        if (df.loc[index, 'C1_distance^2'] > df.loc[index, 'C2_distance^2']):
            new_C1 = np.append(new_C1, index)
        else: 
            new_C2 = np.append(new_C2, index)
    
    C1 = new_C1
    C2 = new_C2
```

```python
# printing clusters

print("final cluster 1: ",new_C1)
print("final cluster 2: ",new_C2)
```

```python
# printing targets of cluster 1

for i in C1:
    print(df.loc[df.index[i], 'sum>10'], end=" ")
    df.loc[df.index[i], 'cluster'] = 0
```

```python
# printing targets of cluster 2

for i in C2:
    print(df.loc[df.index[i], 'sum>10'], end=" ")
    df.loc[df.index[i], 'cluster'] = 1
```
