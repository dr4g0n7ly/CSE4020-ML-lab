```python
# import libraries
import numpy as np
import pandas as pd

# creating the Numpy array
array = np.array([
    [1, 1, 0], 
    [2, 4, 0], 
    [3, 9, 1], 
    [4, 6, 1], 
    [5, 5, 1], 
    [6, 3, 0],              
    [7, 4, 1],
    [8, 9, 1],
    [1, 5, 0], 
    [4, 3, 0],              
    [7, 1, 0],
    [3, 9, 1],
    [0, 0, 0],
    [1, 0, 0],
    [2, 2, 0]
])

# creating a list of column names
column_values = ['x', 'y', 'class']

# creating dataframe
df = pd.DataFrame(
    data = array,
    columns = column_values
)

# displaying the dataframe
df
```

```python
# test values
xt = 4
yt = 7

# length of data
n = df.shape[0]

# finding euclidian distances
for i in range(n):
    x = df.loc[df.index[i], 'x']
    y = df.loc[df.index[i], 'y']
    
    d2 = (xt - x)**2 + (yt - y)**2
    d = d2**0.5
    
    df.loc[df.index[i], 'dist'] = d
    
df
```

```python
# sorting distances
df.sort_values(by=['dist'], inplace=True)
df
```

```python
# finding k nearest neighbours
k = 3
df = df.head(k)
df
```

```python
# finding class (average)

ans = df['class'].mean()
print("test x: " + str(xt))
print("test y: " + str(yt))
if (ans > 0.5):
    print("PREDICTION: greater than 10")
else:
    print("PREDICTION: less than 10")
```
